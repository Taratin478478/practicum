```python
df.publishedAt = pd.to_datetime(df.publishedAt)
```

```python
df.publishedAt.dt.date
```

```python
import matplotlib.pyplot as plt

count = df.drop_duplicates(subset=['thumbnail_link']).groupby(df["publishedAt"].dt.date)["thumbnail_link"].count()
plt.figure(figsize=(14, 8))
plt.plot(count.index, count.values)
plt.xticks(rotation=45, ha="right")
plt.title("-    07.11.2022   ")
plt.xlabel(" ")
plt.ylabel("  ")
```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
count = df.groupby("country")["video_id"].count().sort_values()
plt.title("-    07.11.2022  ")
plt.xlabel("")
plt.ylabel("  ")
plt.bar(count.index, count.values)
```
